dropping method

```python
#drop missing value
credit_card_data=credit_card_data.dropna(how='any')
```

```python
credit_card_data.isnull().sum()
```

```python
#distribiution of legit transaction's fraudulent transactions
credit_card_data['Class'].value_counts()
```

this dataset is highly unbalanced
0--> normal transactions
1-->fraudulent transaction

```python
#separating  the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

under-Sampling

build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

number of Fraudulent transactions-->103

```python
legit_sample=legit.sample(n=103)
```

concatenating two data samples

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting the data into features and targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

split the data into training data and testing data

```python
#features are stored in x_train and corresponding labels are stored in y_train
#80% of data is in x_train and 20%in x_test
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

model training

Logistic regression

1 model=LogisticRegression()

```python
model=LogisticRegression(max_iter=5000)
```

```python
#training the logistic Regression Model with training data # Increase max_iter to 1000
model.fit(X_train,Y_train)
```

Model Evaluate

Accuracy Score

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy score:',test_data_accuracy)
```
